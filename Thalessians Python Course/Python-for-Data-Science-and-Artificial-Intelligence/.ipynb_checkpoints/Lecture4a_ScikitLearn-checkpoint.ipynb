```python
training_n = 10
training_x = x[0:training_n, 1:]
training_y = y[0:training_n, :]
test_x = x[training_n:, 1:]
test_y = y[training_n:, :]
```

```python
scaler = StandardScaler()
scaler.fit(training_x)
training_x = scaler.transform(training_x)
test_x = scaler.transform(test_x)
```

```python
model = neural_network.MLPRegressor()
model.fit(training_x, training_y.ravel())
```

```python
training_yhat = model.predict(training_x); training_yhat
```

```python
{'MSE': mean_squared_error(training_y, training_yhat), 'R2': r2_score(training_y, training_yhat)}
```

```python
test_yhat = model.predict(test_x); test_yhat
```

```python
{'MSE': mean_squared_error(test_y, test_yhat), 'R2': r2_score(test_y, test_yhat)}
```

```python
pylab.plot(test_y, test_yhat, 'x')
```
