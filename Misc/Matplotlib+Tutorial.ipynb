# Matplotlib

Matplotlib is a graphical library for Python which is compatible with Numpy arrays and Pandas dataframes and was created to replicate Matlab graphing capabilities.
Outputs from matplotlib can be directly displayed in Jupyter notebook!


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Lets create a simple array for x and y axis

```python
x = np.linspace(0,0.5,15)
y = x ** 3
```

```python
plt.xlabel("Label x")
plt.ylabel("Label y")
plt.title("Title")
plt.plot(x,y,'g')
```

# Plotting Using Matplotlib (The Right Way)

```python
x = np.array(range(0,5))
y = x ** 2

print(x)
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([1, 1, 1, 1]) # left, bottom, width, height (range 0 to 1) - Position of the axis
# Plot on that set of axes

#axes.set_xticks(x)
#axes.set_yticks(y)
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

# Creating Multiple Axes

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2') # can accept LaTex
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

# Plotting multiple lines

```python
fig, axes = plt.subplots() #subplots creates figure with axes already made. Use tuple unpacking for assignment

axes.plot(x, x**2, label="$x^2$")
axes.plot(x, x**3, label="$x^3$") #label arguement to supply legends

```

# Adding Legends

```python
axes.legend() #places legend on axes
fig
```

```python
axes.legend(loc=1) # upper right corner
axes.legend(loc=2) # upper left corner
axes.legend(loc=3) # lower left corner
axes.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
axes.legend(loc=0) # let matplotlib decide the optimal location


fig
```

# Adding Labels to Lines

```python
axes.text(0.2,0.08,"$(x^2)$", fontsize = 20, color= "blue")
axes.text(0.35,0.07,"$(x^3)$", fontsize= 20, color= "green")

axes.set_ylabel('$y =x^2$',fontdict= {'fontsize':20})
axes.set_xlabel('$x$',fontdict= {'fontsize':20})
fig
```

# Setting Custom Axes Range

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes") #x and y limit = max(x), max(y)

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 0.10])
axes[2].set_xlim([0, 0.2])
axes[2].set_title("custom axes range");
```

# Formatting Lines

```python
#create a blank canvas
fig, axes = plt.subplots()

#add an axis

X = np.linspace(0,0.5,10)

#some cool formatting
axes.plot(x,y,'r--',label='y')
axes.plot(x,y**2,'b-.',label = 'y^2')
axes.plot(x,2*y,'g.-', label = '2y')
axes.plot(x,0.5*y,'y-*', label = '0.5y')
axes.legend(loc=0)
```

We can optionally use the __color__ parameter to provide RBG hexcodes or colour strings.

Matplotlib also provides the ability to set the opacity of lines with the __alpha__


```python
fig, axes = plt.subplots()

axes.plot(x,y,color='#8B008B',label='y')
axes.plot(x,y**2,color = '#FF8C00',label = 'y^2')
axes.plot(x,0.5*y,'blue', label = '0.5y', alpha = 0.5)
axes.legend(loc =0)
```

# More lines and markers

```python
fig, ax = plt.subplots(figsize=(12,6)) #figsize changes the size of the graph

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

# 3D Plots

```python
#Creating some data

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(14,6))

# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1,cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```
